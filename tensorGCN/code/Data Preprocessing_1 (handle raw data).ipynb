# Movies (Items)

```python
type_list = [
    "Action", "Adventure", "Animation",
    "Children's", "Comedy", "Crime",
    "Documentary", "Drama",
    "Fantasy", "Film-Noir"
    "Horror", "Musical", "Mystery",
    "Romance", "Sci-Fi", "Thriller",
    "War", "Western"]
```

```python
filename = "movies.dat"
f = open(filename, encoding="ISO-8859-1")
```

```python
movie_info = []
for line in f:
    info = line.split("::")
    (movie_id, name, types) = info
    movie_id = int(movie_id)
    types = types.strip()
    # tmp = types.split("|")
    tmp_info = {"movie_id": movie_id, "name": name}
    for i in type_list:
        if i in types.split("|"):
            tmp_info[i] = 1
        else:
            tmp_info[i] = 0
    movie_info.append(tmp_info)
    print(line.strip())
```

```python
movie_info
```

```python
movie_df = pd.DataFrame(movie_info)
movie_df.shape
```

```python
movie_df.to_csv("movies.csv")
```

```python
# 打印出缺失信息的电影ID
for i in range(1, 3953):
    if i not in list(movie_df["movie_id"]):
        print(i)
```

# Rating

```python
filename = "ratings.dat"
f = open(filename, encoding="ISO-8859-1")
time_list = []
for line in f:
    time_list.append(int(line.split("::")[-1]))
```

```python
min(time_list), max(time_list)
```

```python
time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(min(time_list))), \
time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(max(time_list)))
```

```python
tmp = max(time_list)
# months_dict = defaultdict()
datetime.fromtimestamp(tmp).year - 2000
```

```python
filename = "ratings.dat"
f = open(filename, encoding="ISO-8859-1")

rating_info = []
for line in f:
    info = line.split("::")
    (user_id, movie_id, rating, timestamp) = info
    user_id = int(user_id)
    movie_id = int(movie_id)
    rating = int(rating)
    timestamp = int(timestamp)
    year_ = datetime.fromtimestamp(timestamp).year - 2000
    month_ = datetime.fromtimestamp(timestamp).month - 4
    month_ = month_ + year_ * 12
    
    tmp_info = {"UserID": user_id, "MovieID": movie_id,
               "Rating": rating, "TimeStamp": timestamp,
               "Month": month_}
    rating_info.append(tmp_info)
    
rating_df = pd.DataFrame(rating_info)
rating_df.to_csv("ratings.csv")
```
